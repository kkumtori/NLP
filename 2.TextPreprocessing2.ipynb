## 2-5. 정규 표현식(Regular Expression)


### 2-5-1. 정규 표현식 문법과 모듈 함수  
특수 문자
+ `.` : 임의의 문자('\n'은 제외) 1개
+ `?` : 앞의 문자 0개 or 1개  
+ `*` : 앞의 문자 0개 이상
+ `+` : 앞의 문자 1개 이상
+ `^` : 뒤의 문자열로 문자열 시작
+ `$` : 앞의 문자열로 문자열 끝  
+ `{숫자}` : 숫자만큼 반복
+ `{숫자1, 숫자2}` : [숫자1,숫자2] 만큼 반복  
+ `{숫자, }` : 숫자 이상만큼 반복  
+ `[문자]` : [amk] a or m or k
+ `[^문자]` : 문자 제외 문자 매치
+ `|` : [A|B] A 또는 B  
  
탈출 문자 
+ `\\\` : 역슬래쉬 자체
+ `\\d` : 모든 숫자, [0-9] 
+ `\\D` : 숫자제외 모든 문자, [^0-9]
+ `\\s` : 공백, [ \t\n\r\f\v]
+ `\\S` : 공백제외 모든 문자, [^\t\n\r\f\v]
+ `\\w` : 모든 문자 or 숫자, [a-zA-Z0-9]
+ `\\W` : 문자 또는 숫자가 아닌 문자 [^a-zA-Z0-9]

### 2-5-5. 정규 표현식을 이용한  토큰화  
`RegexpTokenizer` : 정규 표현식을 사용한 단어 토큰화 도구  
+ RegexpTokenizer() 안에 하나의 토큰으로 규정하기 원하는 정규식 넣어서 토큰화 수행

```python
from nltk.tokenize import RegexpTokenizer

text="Don't be fooled by the dark sounding name, \
Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"
```

```python
# tokenizer1 : 문자 or 숫자 1개 이상 -> 하나의 토큰으로
tokenizer1=RegexpTokenizer('\w+')
print(tokenizer1.tokenize(text))
```

```python
# tokenizer2 : 공백 기준 토큰화
tokenizer2=RegexpTokenizer('\s+',gaps=True) # gaps=True: 정규식을 기준으로
print(tokenizer2.tokenize(text))
```

## 2-6. 정수 인코딩 (Integer Encoding)  
5000개의 단어가 등장한다면   
-> 빈도수로 정렬  
-> 1~5000번 mapping

### 2-6-1. Integer Encoding

```python
import nltk
nltk.download('punkt')
```

```python
# dictionary 사용
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
raw_text = "A barber is a person. \
a barber is good person.\
 a barber is huge person. \
 he Knew A Secret! \
 The Secret He Kept is huge secret. \
 Huge secret. \
 His barber kept his word. \
 a barber kept his word. \
 His barber kept his secret. \
 But keeping and keeping such a huge secret to himself was driving the barber crazy. \
 the barber went up a huge mountain."
```

```python
# 문장 토큰화 
s_tokens=sent_tokenize(raw_text)
print(s_tokens)
```

```python
# vocab 딕셔너리 생성
vocab={}
preproc_sentences=[]
stop_words=set(stopwords.words('english'))

for s in s_tokens:
    w_tokens=word_tokenize(s) # 단어 토큰화
    
    result = []
    for w in w_tokens:
        w=w.lower() # 소문자로 통일
        if w not in stop_words: # 불용어 제거
            if len(w)>2: # 길이 기준으로 Cleaning
                result.append(w)
                if w not in vocab:
                    vocab[w]=0
                vocab[w]+=1

    preproc_sentences.append(result)
```

```python
print(preproc_sentences)
```

```python
print('단어집합 : ',vocab)
```

```python
# 딕셔너리 정렬
vocab_sorted=sorted(vocab.items(),key=lambda x:x[1],reverse=True)
print(vocab_sorted)
```

```python
# mapping
word2index={}

i=0
for (word,freq) in vocab_sorted: 
    if freq>1:
        i=i+1
        word2index[word]=i
```

```python
print(word2index)
```

```python
# 빈도수 상위 5개 단어만 남기고 삭제
size=5

## 희귀 단어 리스트
rare_words=[word for (word,rank) in word2index.items() if rank>size]
print(rare_words)

## 나머지 삭제
for word in rare_words:
    del word2index[word]
print(word2index)
```

```python
# preproc_sentences를 인코딩하기
print(preproc_sentences)
```

```python
word2index['OOV']=len(word2index)+1
print(word2index)
```

```python
encoded_sentences=[]
for s in preproc_sentences:
    temp=[]
    for w in s:
        try:
            temp.append(word2index[w])
        except KeyError:
            temp.append(word2index['OOV'])
    encoded_sentences.append(temp)
```

```python
print(encoded_sentences)
```
