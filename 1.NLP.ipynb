### Numpy

```python
import numpy as np
```

```python
# 1차원 배열
vec=np.array([1,2,3,4,5])
print(vec)
```

```python
# 2차원 배열
mat=np.array([[10,20,30],[60,70,80]])
print(mat)
```

```python
print('vec의 타입 : ',type(vec))
print('mat의 타입 : ',type(mat))
```

+ `ndim` : 차원(축의 개수)
+ `shape` : 크기

```python
print('vec.ndim : ',vec.ndim)
print('vec.shape : ', vec.shape)
```

```python
print('mat.ndim : ',mat.ndim)
print('mat.shape : ',mat.shape)
```

```python
# ndarray 초기화
## 0으로 채워진 2x3 배열 생성
zero_mat=np.zeros((2,3))
print(zero_mat)
```

```python
## 1로 채워진 2x3 배열 생성
one_mat=np.ones((2,3))
print(one_mat)
```

```python
## 사용자 지정값으로 채워진 2x3 배열 생성
const_mat=np.full((2,3),7)
print(const_mat)
```

```python
## Identity Matrix 
diag_mat=np.eye(3)
print(diag_mat)
```

```python
## 랜덤값으로 채워진 배열
random_mat=np.random.random((2,2))
print(random_mat)
```

```python
# np.arrange()
range_vec=np.arange(10) ## [0,10)
print(range_vec)
```

```python
n=2
range_vec2=np.arange(1,10,n)
print(range_vec2)
```

```python
# np.reshape()
mat=np.array(np.arange(30))
print(mat) # 30x1
print(mat.shape)
mat2=mat.reshape(6,5) 
print(mat2) # 6x5
print(mat2.shape)
```

```python
# numpy 슬라이싱
mat=np.array([[1,2,3],[4,5,6]])
print(mat) # 2x3
```

```python
mat[0,:] # 0번째 row
```

```python
mat[:,1] # 1번째 col
```

```python
# 정수 인덱싱
mat=np.array([[1,2],[3,4],[5,6]])
print(mat)
```

```python
## 1row 0col element
print(mat[1,0])

## (2,0), (1,1) element
print(mat[[2,1],[0,1]])
```

```python
# Numpy 연산
x=np.array([1,2,3])
y=np.array([4,5,6])

print(x+y)
print(x-y)
print(x*y)
print(x/y)
```

요소별 연산 수행



```python
print(np.add(x,y))
print(np.subtract(x,y))
print(np.multiply(x,y))
print(np.divide(x,y))
```

```python
# 행렬곱
mat1=np.array([[1,2],[3,4]])
mat2=np.array([[5,6],[7,8]])

np.dot(mat1,mat2)
```
